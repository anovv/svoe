```python
def compute_vwap(df, key):
    q = df['amount']
    p = df['price']
    vwap = np.sum(p * q) / np.sum(q)
    df[key] = vwap
    return df
```

```python
# time bars

df['ts'] = pd.to_datetime(df['timestamp'], unit='s')
# df.set_index('ts', inplace=True)
df = df.set_index('ts')
time_freq = '1Min'
key_time_vwap = 'time_vwap_' + time_freq
# df_vwap = df.groupby(pd.Grouper(freq=time_freq)).apply(compute_vwap, key=key_time_vwap)
# df_vwap
df_time_grp = df.groupby(pd.Grouper(freq=time_freq))
num_time_bars = len(df_time_grp)
df_time_vwap = df_time_grp.apply(compute_vwap, key=key_time_vwap)
```

```python
df_time_vwap.plot(y=['price', key_time_vwap])
```

```python
# tick bars

total_ticks = len(df)
tick_freq = total_ticks / num_time_bars # tick per bar
tick_freq = round(tick_freq, -3) # round to the nearest thousand

key_tick_vwap = 'tick_vwap_' + str(int(tick_freq))

df_tick_grp = df.reset_index().assign(grpId=lambda row: row.index // tick_freq)
df_tick_vwap =  df_tick_grp.groupby('grpId').apply(compute_vwap, key=key_tick_vwap)
df_tick_vwap.set_index('ts', inplace=True)
```

```python
df_tick_vwap.plot(y=['price', key_tick_vwap])
```

```python
# volume bars

df_cum_vol = df.assign(cum_vol=df['amount'].cumsum()) 
total_vol = df_cum_vol.cum_vol.values[-1]
vol_freq = total_vol / num_time_bars # vol per bar
vol_freq = round(vol_freq, -2) # round to the nearest hundred

key_vol_vwap = 'vol_vwap_' + str(int(vol_freq))

df_vol_grp = df_cum_vol.assign(grpId=lambda row: row.cum_vol // vol_freq)
df_vol_vwap =  df_vol_grp.groupby('grpId').apply(compute_vwap, key=key_vol_vwap)
df_vol_vwap
```

```python
df_vol_vwap.plot(y=['price', key_vol_vwap])
```

```python
# dollar bars

df['doll'] = df['amount'] * df['price']
df_cum_doll = df.assign(cum_doll=(df['doll']).cumsum()) 
total_doll = df_cum_doll.cum_doll.values[-1]
doll_freq = total_doll / num_time_bars # dollar volume per bar
doll_freq = round(doll_freq, -2) # round to the nearest hundred

key_doll_vwap = 'doll_vwap_' + str(int(doll_freq)) # TODO set fixed dollar value

df_doll_grp = df_cum_doll.assign(grpId=lambda row: row.cum_doll // doll_freq)
df_doll_vwap =  df_doll_grp.groupby('grpId').apply(compute_vwap, key=key_doll_vwap)
df_doll_vwap
```

```python
df_doll_vwap.plot(y=['price', key_doll_vwap])
```

```python
%%time
# tick direction
df = df.reset_index()
df['price_diff'] = df['price'].diff().fillna(0)

prev_direction = 1
for index, row in df.iterrows():
    price_diff = row['price_diff']
    if price_diff == 0.0:
        df.at[index, 'tick_direction'] = prev_direction
    else:
        direction = 1 if price_diff > 0.0 else -1
        df.at[index, 'tick_direction'] = direction
        prev_direction = direction

df = df.set_index('ts')
df
```

```python
# calc signed flow
df['signed_doll'] = df['doll'] * df['tick_direction']
df
```

```python
# dollar imbalance bars # TODO compute tick/vol imbalance as well

# From de Lopez: "As we will see in Chapter 19, market microstructure theories
# confer special importance to the persistence of imbalanced signed volumes, as
# that phenomenon is associated with the presence of informed traders. By synchronizing
# sampling with the arrival of informed traders, we may be able to make decisions
# before prices reach a new equilibrium level"

# Implementation is from https://towardsdatascience.com/financial-machine-learning-part-0-bars-745897d4e4ba

# TODO explore hyperparam tuning 
# https://towardsdatascience.com/information-driven-bars-for-financial-machine-learning-imbalance-bars-dda9233058f0

# TODO numbarize

# TODO explore
# https://github.com/BlackArbsCEO/Adv_Fin_ML_Exercises/issues/1
# https://github.com/BlackArbsCEO/Adv_Fin_ML_Exercises/issues/1#issuecomment-631385708
# https://github.com/BlackArbsCEO/Adv_Fin_ML_Exercises/issues/1#issuecomment-705220622

# TODO use Kalman filter instead of EWMA
init_abs_expected_signed_doll = np.abs(df['signed_doll'].mean())
init_expected_num_ticks = 500000 # 500000 ticks to warm up # TODO test this!!

def compute_doll_imb_bars(df, init_expected_num_ticks, init_abs_expected_signed_doll):
    nums_ticks, bar_start_indeces = [], []
    prev_index, expected_num_ticks, abs_expected_signed_doll  = 0, init_expected_num_ticks, init_abs_expected_signed_doll
    
    n = df['signed_doll'].shape[0]
    signed_doll_vals = df['signed_doll'].values.astype(np.float64)
    
    # theta is cumulative signed dollar volume in sample (inside the bar)
    # threshold determines if we should start sampling when absolute theta value is above it
    abs_thetas, thresholds = np.zeros(n), np.zeros(n)
    abs_thetas[0], cur_theta = np.abs(signed_doll_vals[0]), signed_doll_vals[0]
    for i in tqdm(range(1, n)):
        cur_theta += signed_doll_vals[i]
        abs_theta = np.abs(cur_theta)
        abs_thetas[i] = abs_theta
        
        threshold = expected_num_ticks * abs_expected_signed_doll
        thresholds[i] = threshold
        
        if abs_theta >= threshold:
            cur_theta = 0
            nums_ticks.append(np.float64(i - prev_index))
            bar_start_indeces.append(i)
            prev_index = i
            expected_num_ticks = de.ewma_infinite_hist(np.array(nums_ticks), window=np.int64(len(nums_ticks)))[-1] # TODO use fixed window/span here?
            abs_expected_signed_doll = np.abs(de.ewma_infinite_hist(signed_doll_vals[:i], window=np.int64(init_expected_num_ticks * 3))[-1] ) # window of 3 bars
    
    return nums_ticks, abs_thetas, thresholds, bar_start_indeces

```

```python
%%time
nums_ticks, abs_thetas, thresholds, bar_start_indeces = compute_doll_imb_bars(df, init_expected_num_ticks, init_abs_expected_signed_doll)
```

```python
plt.plot(abs_thetas)
plt.plot(thresholds)
plt.show()
```

```python
n = df['signed_doll'].shape[0]
i_iter = iter(bar_start_indeces + [n])
i_cur = i_iter.__next__()
grpId = np.zeros(n)
for i in range(1, n):
    if i <= i_cur:
        grpId[i] = grpId[i-1]
    else:
        grpId[i] = grpId[i-1] + 1
        i_cur = i_iter.__next__()
```

```python
df_doll_imb_grp = df.assign(grpId = grpId)

key_doll_imb_vwap = 'doll_imb_vwap' # to do pass hyperparams

df_doll_imb_vwap = df_doll_imb_grp.groupby('grpId').apply(compute_vwap, key=key_doll_imb_vwap)
df_doll_imb_vwap
```

```python
# dollar run bars # TODO compute tick/vol run as well

# From de Lopez: "Large traders will sweep the order book, use
# iceberg orders, or slice a parent order into multiple children, all of which leave a
# trace of runs in the {bt}t=1,,T sequence. For this reason, it can be useful to monitor
# the sequence of buys in the overall volume, and take samples when that sequence
# diverges from our expectations"

# Implementation is similar to compute_doll_imb_bars

# sample implementation:
# https://github.com/BlackArbsCEO/Adv_Fin_ML_Exercises/issues/1#issuecomment-441312197

# TODO numbarize
init_expected_doll_buy = None # TODO
init_expected_doll_sell = None # TODO
init_expected_proportion_buys = None
init_expected_num_ticks = 30000 # 30000 ticks to warm up # TODO test this!!

def compute_doll_run_bars(df, init_expected_num_ticks, init_abs_expected_signed_doll):
    nums_ticks, bar_start_indeces, proportions_buys, doll_buys, doll_sells = [], [], [], [], []
    prev_index, expected_num_ticks, expected_doll_buy, expected_doll_sell = 0, init_expected_num_ticks, init_expected_doll_buy, init_expected_doll_sell
    
    n = df['signed_doll'].shape[0]
    signed_doll_vals = df['signed_doll'].values.astype(np.float64)
    
    # theta is cumulative signed dollar volume in sample (inside the bar)
    # threshold determines if we should start sampling when theta value is above it
    thetas, thresholds = np.zeros(n), np.zeros(n)
    thetas[0], cur_theta = np.abs(signed_doll_vals[0]), signed_doll_vals[0]
    for i in tqdm(range(1, n)):
        cur_theta += signed_doll_vals[i]
        thetas[i] = abs_theta
        
        threshold = expected_num_ticks * abs_expected_signed_doll
        thresholds[i] = threshold
        
        if abs_theta >= threshold:
            cur_theta = 0
            nums_ticks.append(np.float64(i - prev_index))
            bar_start_indeces.append(i)
            prev_index = i
            # TODO use fast ewma
            expected_num_ticks = de.ewma(np.array(nums_ticks), window=np.int64(len(nums_ticks)))[-1] # TODO use fixed window/span here?, #TODO uase mean() instead of last value?
            abs_expected_signed_doll = np.abs(de.ewma(signed_doll_vals[:i], window=np.int64(init_expected_num_ticks * 3))[-1] ) # window of 3 bars
    
    return nums_ticks, abs_thetas, thresholds, bar_start_indeces

```
